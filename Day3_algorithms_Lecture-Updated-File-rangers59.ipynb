## Two Pointers

#### Syntax

```python


def two_point(my_list):
    #create "pointers" for the list
    left = 0
    right = len(my_list) - 1
    
    print(f'left: {left} right: {right}')
    
    while left <= right:
        print(f'sample_list in loop: {my_list}')
        my_list[left], my_list[right] = my_list[right], my_list[left]
        left += 1
        right -= 1
    return my_list

sample_list = [10,11,12,13,14]

two_point(sample_list)
    
    
print(f'sample after: {sample_list}')
```

# Sorting Algorithms

#### Bubble Sort

Worst Case: O(n^2) Time - O(1) Space

```python
list_a = [22, 55, 88, 44, 1, 100, 34, 66]

#helper function to swap

def swap(arr, i, j):
    arr[i], arr[j] = arr[j], arr[i]

print(f'list_a before: {list_a}')
    
swap_(list_a, 2, 3)

print(f'list_a after: {list_a}')

def bubble(arr):
    is_sorted = False
    while not is_sorted:
        is_sorted = True
        for num in range(len(arr)-1):
            if arr[num]> arr[num + 1]:
                swap(arr, num, num + 1)
                is_sorted = False
    print(f'arr after bubble sort: {arr}')            
    return arr

bubble(list_a)
```

##### Insertion Sort

Worst Case: O(n^2) time - O(1)space

```python
list_a = [22, 55, 88, 44, 1, 100, 34, 66]

#helper function to swap

def swap(arr, i, j):
    arr[i], arr[j] = arr[j], arr[i]
    
def insert(arr):
    for i in range(1,len(arr)):
        while i > 0 and arr[i] < arr[i - 1]:
            swap(arr, i, i - 1)
            i -= 1
        return arr
    
insert(list_a)
```

```python
a_text = 'In computing, a hash table hash map is a data structure which implements an associative array abstract data type, a structure that can map keys to values. A hash table uses a hash function to compute an index into an array of buckets or slots from which the desired value can be found'

def word_count(str):
    counts = dict()
    words = str.split()

    for word in words:
        if word in counts:
            counts[word] += 1
        else:
            counts[word] = 1

    return counts

print(word_count(a_text))
```
